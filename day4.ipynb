```
df['brand_cat'] = df['brand'].factorize()[0]
```

```
feats = ['brand_cat']
X=df[ feats ].values
Y=df['prices_amountmin'].values

model = DecisionTreeRegressor(max_depth=5)
import sklearn
sklearn.metrics.SCORERS.keys()

scores=cross_val_score(model, X, Y,scoring = 'neg_median_absolute_error')
np.mean(scores), np.std(scores)
```

```
scores=cross_val_score(model, X, Y,scoring = 'neg_median_absolute_error')
np.mean(scores), np.std(scores)
```

```
def run_model(feats):
 
  X=df[ feats ].values
  Y=df['prices_amountmin'].values

  model = DecisionTreeRegressor(max_depth=5)

  scores=cross_val_score(model, X, Y,scoring = 'neg_median_absolute_error')
  return np.mean(scores), np.std(scores)
```

```
run_model(['brand_cat'])
```

```
df['manufacturer_cat'] = df['manufacturer'].factorize()[0]
```

```
run_model(['manufacturer_cat'])
```

```
run_model(['manufacturer_cat','brand_cat'])
```

```
 df['flavors_cat'] = df['flavors'].factorize()[0]
```

```
run_model(['flavors_cat','brand_cat','manufacturer_cat'])
```
